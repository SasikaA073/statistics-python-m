```python
from math import comb
from math import perm
import matplotlib.pyplot as plt
```

```python

def plot_bernoulli_distibution(n,s, img_save = False):

    n = int(n)
    x = []
    y = []
    for i in range(0,n):
        prob = comb(n,i)*((s)**i)*((1-s)**(n-i))
        # print(f"P({i}) = {prob}")
        x.append(i)
        y.append(prob)

    # start a new figure
    plt.figure()
    
    # Plot only the points
    plt.grid(True)
    plt.scatter(x,y)
    plt.title(f"n = {n}, s = {s}")

    # Plot the line
    plt.plot(x,y)


    if img_save:
        # save image
        plt.savefig(f"./n_10/bernoulli_distibution_n_{n}_s_{s}.png")
```

```python
count = 0
increment = 0.05
while count <= 1 + increment:
    plot_bernoulli_distibution(10,count)
    count += increment
```

```python
plot_bernoulli_distibution(4,0.05)
```
